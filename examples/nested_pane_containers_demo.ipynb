# Nested Pane Containers Demo

This notebook demonstrates how to create nested pane containers, where one pane container is placed inside another. Each container maintains its own independent tabs and state.

```python
# Standard Library Imports
import sys
from pathlib import Path

# Add the parent directory to the path so we can import the tts_html_utils package
sys.path.insert(0, str(Path().resolve().parent.parent))

# Teamtools Studio Imports
from tts_html_utils.core.components.structure import PaneContainer
from tts_html_utils.core.components.table import Table
from tts_html_utils.core.components.misc import Div, Heading, Paragraph
```

## Helper Functions

Let's create some helper functions to generate demo content.

```python
def create_demo_table(title, rows=3, cols=3):
    """Create a simple demo table with the given title."""
    headers = [f"Column {i+1}" for i in range(cols)]
    data = [[f"{title} Row {i+1}, Col {j+1}" for j in range(cols)] for i in range(rows)]
    return Table(headers=headers, data=data, title=title)


def create_demo_content(title):
    """Create some demo content with the given title."""
    return Div([
        Heading(title, level=3),
        Paragraph(f"This is some demo content for {title}."),
        Paragraph("Lorem ipsum dolor sit amet, consectetur adipiscing elit. Sed do eiusmod tempor incididunt ut labore et dolore magna aliqua.")
    ])
```

## Creating Nested Pane Containers

Now let's create nested pane containers to demonstrate how they work.

```python
# Create the outer container with a custom ID
outer_container = PaneContainer(id="outer-container")

# Create the first inner container with a custom ID
inner_container1 = PaneContainer(id="inner-container1")

# Add panes to the first inner container
inner_container1.add_pane(create_demo_table("Inner Table 1"), "Inner Tab 1")
inner_container1.add_pane(create_demo_content("Inner Content 1"), "Inner Tab 2")
inner_container1.add_pane(create_demo_table("Inner Table 2", rows=2, cols=4), "Inner Tab 3")

# Create the second inner container with an auto-generated ID
inner_container2 = PaneContainer()

# Add panes to the second inner container
inner_container2.add_pane(create_demo_content("Another Inner Content"), "Another Inner Tab 1")
inner_container2.add_pane(create_demo_table("Another Inner Table", rows=4, cols=2), "Another Inner Tab 2")

# Add the inner containers and other content to the outer container
outer_container.add_pane(create_demo_content("Outer Content 1"), "Outer Tab 1")
outer_container.add_pane(inner_container1, "Nested Container 1")
outer_container.add_pane(create_demo_table("Outer Table", rows=5, cols=3), "Outer Tab 3")
outer_container.add_pane(inner_container2, "Nested Container 2")
```

## Displaying the Nested Pane Containers

Now let's display the nested pane containers.

```python
# Display the outer container
from IPython.display import display, HTML
display(HTML(outer_container.render()))
```

## Explanation

In this demo, we created:

1. An outer pane container with ID "outer-container"
2. Two inner pane containers:
   - One with ID "inner-container1"
   - One with an auto-generated ID

Each container has its own tabs and content. The outer container has tabs for:
- "Outer Tab 1" (simple content)
- "Nested Container 1" (contains the first inner container)
- "Outer Tab 3" (a table)
- "Nested Container 2" (contains the second inner container)

The inner containers have their own tabs that work independently of the outer container's tabs.

This demonstrates how pane containers can be nested while maintaining independent functionality.
